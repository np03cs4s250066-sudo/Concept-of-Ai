Problem - 3: Array Operations:


```python
import numpy as np

# Given arrays
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

print("Original arrays:")
print("x =\n", x)
print("y =\n", y)
print("v =", v)
print("w =", w)

# Task 1: Add arrays
add_xy = x + y
print("\n1. x + y:")
print(add_xy)

# Task 2: Subtract arrays
sub_xy = x - y
print("\n2. x - y:")
print(sub_xy)

# Task 3: Multiply with integer
mult_x = x * 3
mult_y = y * 2
print("\n3. x * 3:")
print(mult_x)
print("y * 2:")
print(mult_y)

# Task 4: Square of each element
square_x = x ** 2
square_y = y ** 2
print("\n4. Square of x:")
print(square_x)
print("Square of y:")
print(square_y)

# Task 5: Dot products
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)

print("\n5. Dot products:")
print("v · w =", dot_vw)
print("x · v =", dot_xv)
print("x · y =\n", dot_xy)

# Task 6: Concatenation
concat_row = np.concatenate((x, y), axis=0)  # Along rows
concat_col_vw = np.concatenate((v.reshape(-1,1), w.reshape(-1,1)), axis=1)  # Along columns

# Alternative using vstack and hstack
concat_row_alt = np.vstack((x, y))
concat_col_alt = np.column_stack((v, w))

print("\n6. Concatenation:")
print("x and y along rows:\n", concat_row)
print("v and w along columns:\n", concat_col_vw)

# Task 7: Concatenate x and v (will cause error)
print("\n7. Concatenate x and v:")
try:
    concat_xv = np.concatenate((x, v))
    print("Success:", concat_xv)
except ValueError as e:
    print("Error:", e)
    print("Explanation: Cannot concatenate arrays with different dimensions.")
    print("x shape:", x.shape, "(2D array)")
    print("v shape:", v.shape, "(1D array)")
    print("All input arrays must have the same number of dimensions for concatenation.")
```

Problem - 4: Matrix Operations:

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Original matrices:")
print("A =\n", A)
print("B =\n", B)

print("\n" + "="*50)
print("1. Prove A × A⁻¹ = I")
print("="*50)

# Calculate inverse of A
try:
    A_inv = np.linalg.inv(A)
    print("A⁻¹ =\n", A_inv)

    # Multiply A and A inverse
    product = np.dot(A, A_inv)
    print("\nA × A⁻¹ =\n", product)

    # Create identity matrix for comparison
    I = np.eye(2)
    print("\nIdentity matrix I =\n", I)

    # Check if they are equal (considering floating point precision)
    are_equal = np.allclose(product, I)
    print(f"\nA × A⁻¹ = I: {are_equal}")

except np.linalg.LinAlgError:
    print("Matrix A is singular, cannot compute inverse")

print("\n" + "="*50)
print("2. Prove AB ≠ BA")
print("="*50)

# Calculate AB
AB = np.dot(A, B)
print("AB =\n", AB)

# Calculate BA
BA = np.dot(B, A)
print("\nBA =\n", BA)

# Check if they are equal
are_equal = np.array_equal(AB, BA)
print(f"\nAB = BA: {are_equal}")
print(f"Therefore, AB ≠ BA: {not are_equal}")

# Show the difference
difference = AB - BA
print(f"\nDifference (AB - BA) =\n", difference)

print("\n" + "="*50)
print("3. Prove (AB)ᵀ = BᵀAᵀ")
print("="*50)

# Calculate (AB)ᵀ
AB_transpose = AB.T
print("(AB)ᵀ =\n", AB_transpose)

# Calculate BᵀAᵀ
B_transpose = B.T
A_transpose = A.T
BTA_transpose = np.dot(B_transpose, A_transpose)
print("\nBᵀAᵀ =\n", BTA_transpose)

# Check if they are equal
are_equal = np.array_equal(AB_transpose, BTA_transpose)
print(f"\n(AB)ᵀ = BᵀAᵀ: {are_equal}")

# Show they are identical
print(f"\nAre the matrices identical? {np.array_equal(AB_transpose, BTA_transpose)}")
```

```python
import numpy as np

print("Solving System of Linear Equations using Inverse Method")
print("_" * 55)
print("Equations:")
print("2x - 3y +  z = -1")
print(" x -  y + 2z = -3")
print("3x +  y -  z =  9")

# Step 1: Matrix representation
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([[-1],
              [-3],
              [9]])

print("\nStep 1: Matrix Representation")
print("A =")
print(A)
print("\nB =")
print(B)

# Step 2: Check invertibility
print("\nStep 2: Check Invertibility")
det_A = np.linalg.det(A)
print(f"det(A) = {det_A:.2f}")

if abs(det_A) < 1e-10:
    print("Matrix is singular. Cannot use inverse method.")
else:
    print("Matrix is invertible. Proceeding with inverse method.")

    # Step 3: Find inverse
    print("\nStep 3: Find Inverse of A")
    A_inv = np.linalg.inv(A)
    print("A⁻¹ =")
    print(A_inv)

    # Step 4: Solve X = A⁻¹B
    print("\nStep 4: Solve X = A⁻¹B")
    X = np.dot(A_inv, B)
    print("X =")
    print(X)

    x, y, z = X.flatten()
    print(f"\nFinal Solution:")
    print(f"x = {x:.2f}")
    print(f"y = {y:.2f}")
    print(f"z = {z:.2f}")

    # Step 5: Verification
    print("\nStep 5: Verification")
    eq1 = 2*x - 3*y + z
    eq2 = x - y + 2*z
    eq3 = 3*x + y - z

    print(f"Equation 1: 2({x:.2f}) - 3({y:.2f}) + ({z:.2f}) = {eq1:.2f}")
    print(f"Equation 2: {x:.2f} - {y:.2f} + 2({z:.2f}) = {eq2:.2f}")
    print(f"Equation 3: 3({x:.2f}) + {y:.2f} - {z:.2f}) = {eq3:.2f}")

    if np.allclose([eq1, eq2, eq3], [-1, -3, 9]):
        print("\n✓ Solution verified successfully!")
    else:
        print("\n✗ Solution verification failed!")

# Alternative: Using np.linalg.solve() for comparison
print("\n" + "_"*55)
print("Alternative: Using np.linalg.solve()")
X_alt = np.linalg.solve(A, B)
x_alt, y_alt, z_alt = X_alt.flatten()
print(f"x = {x_alt:.2f}, y = {y_alt:.2f}, z = {z_alt:.2f}")
```

4.2 Experiment: How Fast is Numpy?

```python
import numpy as np
import time

# Set the sizes
list_size = 1000000
matrix_size = 1000

print("NUMPY VS PYTHON LISTS PERFORMANCE COMPARISON")
print("=" * 50)

# 1. ELEMENT-WISE ADDITION
print("\n1. ELEMENT-WISE ADDITION")

# Python Lists
python_list1 = list(range(list_size))
python_list2 = list(range(list_size, list_size * 2))

start_time = time.time()
result_python_add = [python_list1[i] + python_list2[i] for i in range(list_size)]
python_add_time = time.time() - start_time
print(f"Python Lists Time: {python_add_time:.6f} seconds")

# NumPy Arrays
numpy_array1 = np.arange(list_size)
numpy_array2 = np.arange(list_size, list_size * 2)

start_time = time.time()
result_numpy_add = numpy_array1 + numpy_array2
numpy_add_time = time.time() - start_time
print(f"NumPy Arrays Time: {numpy_add_time:.6f} seconds")
print(f"Speedup: {python_add_time/numpy_add_time:.2f}x")

# 2. ELEMENT-WISE MULTIPLICATION
print("\n2. ELEMENT-WISE MULTIPLICATION")

# Python Lists
start_time = time.time()
result_python_mul = [python_list1[i] * python_list2[i] for i in range(list_size)]
python_mul_time = time.time() - start_time
print(f"Python Lists Time: {python_mul_time:.6f} seconds")

# NumPy Arrays
start_time = time.time()
result_numpy_mul = numpy_array1 * numpy_array2
numpy_mul_time = time.time() - start_time
print(f"NumPy Arrays Time: {numpy_mul_time:.6f} seconds")
print(f"Speedup: {python_mul_time/numpy_mul_time:.2f}x")

# 3. DOT PRODUCT
print("\n3. DOT PRODUCT")

# Python Lists
start_time = time.time()
dot_python = sum(python_list1[i] * python_list2[i] for i in range(list_size))
python_dot_time = time.time() - start_time
print(f"Python Lists Time: {python_dot_time:.6f} seconds")

# NumPy Arrays
start_time = time.time()
dot_numpy = np.dot(numpy_array1, numpy_array2)
numpy_dot_time = time.time() - start_time
print(f"NumPy Arrays Time: {numpy_dot_time:.6f} seconds")
print(f"Speedup: {python_dot_time/numpy_dot_time:.2f}x")

# 4. MATRIX MULTIPLICATION
print("\n4. MATRIX MULTIPLICATION")

# Python Lists (as nested lists)
python_matrix1 = [[i + j for j in range(matrix_size)] for i in range(matrix_size)]
python_matrix2 = [[i - j for j in range(matrix_size)] for i in range(matrix_size)]

start_time = time.time()
result_python_matmul = [[0 for _ in range(matrix_size)] for _ in range(matrix_size)]
for i in range(matrix_size):
    for j in range(matrix_size):
        for k in range(matrix_size):
            result_python_matmul[i][j] += python_matrix1[i][k] * python_matrix2[k][j]
python_matmul_time = time.time() - start_time
print(f"Python Lists Time: {python_matmul_time:.6f} seconds")

# NumPy Arrays
numpy_matrix1 = np.array(python_matrix1)
numpy_matrix2 = np.array(python_matrix2)

start_time = time.time()
result_numpy_matmul = np.matmul(numpy_matrix1, numpy_matrix2)
numpy_matmul_time = time.time() - start_time
print(f"NumPy Arrays Time: {numpy_matmul_time:.6f} seconds")
print(f"Speedup: {python_matmul_time/numpy_matmul_time:.2f}x")

# SUMMARY
print("\n" + "=" * 50)
print("PERFORMANCE SUMMARY")
print("=" * 50)
print(f"{'Operation':<25} {'Python Time':<12} {'NumPy Time':<12} {'Speedup':<10}")
print("-" * 50)
print(f"{'Element-wise Addition':<25} {python_add_time:<12.6f} {numpy_add_time:<12.6f} {python_add_time/numpy_add_time:<10.2f}x")
print(f"{'Element-wise Multiplication':<25} {python_mul_time:<12.6f} {numpy_mul_time:<12.6f} {python_mul_time/numpy_mul_time:<10.2f}x")
print(f"{'Dot Product':<25} {python_dot_time:<12.6f} {numpy_dot_time:<12.6f} {python_dot_time/numpy_dot_time:<10.2f}x")
print(f"{'Matrix Multiplication':<25} {python_matmul_time:<12.6f} {numpy_matmul_time:<12.6f} {python_matmul_time/numpy_matmul_time:<10.2f}x")
```
